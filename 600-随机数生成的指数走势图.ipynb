```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 采用随机数模拟大盘指数

### 完全随机

```python
change = np.random.randn(8,1000)*2
change
```

```python
plt.scatter(range(1000),change[1])
```

```python
data=[[100],[100],[100],[100],[100],[100],[100],[100]]
plt.figure(figsize=(12,8))
for raw in range(8):
    for index  in range(1000):
        data[raw].append(data[raw][index]*(change[raw][index]/100+1))
    plt.plot(data[raw])
```

### 加上无风险利率

```python
change1 = change+0.025
change1
```

```python
plt.scatter(range(1000),change1[1])
```

```python
data=[[100],[100],[100],[100],[100],[100],[100],[100]]
plt.figure(figsize=(12,8))
for raw in range(8):
    for index  in range(1000):
        data[raw].append(data[raw][index]*(change1[raw][index]/100+1))
    plt.plot(data[raw])
```
